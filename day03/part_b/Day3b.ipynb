```python
import numpy as np
import pandas as pd
from string import ascii_lowercase, ascii_uppercase

from google.colab import files
uploaded = files.upload()
```

```python
letters = [i for i in ascii_lowercase + ascii_uppercase]
mapped_letters = {letter: i + 1 for (i, letter) in enumerate(letters)}
```

```python
df = pd.read_csv("input.txt", names = ["Rutsacks"])
```

```python
df2 = df.assign(one_hundred_repeated = np.repeat(np.array(range(0, 100)), 3))
```

```python
def unique_finder(x):
    try:
        return list(set(x.Rutsacks.tolist()[0]).intersection(set(x.Rutsacks.tolist()[1])).intersection(set(x.Rutsacks.tolist()[2])))[0]
    except:
        return 0
```

```python
new_series = df2.groupby('one_hundred_repeated').apply(unique_finder)
```

```python
df3 = pd.DataFrame(new_series, columns = ["Items"]).assign(priority = lambda x: x.Items.map(mapped_letters))
df3
```
