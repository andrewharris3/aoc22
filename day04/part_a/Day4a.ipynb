```python
df = pd.read_csv('input.txt', names = ['elf1', 'elf2'])
```

```python
df2 = df.assign(elf1_parsed = lambda x: x.elf1.str.split("-"), elf2_parsed = lambda x: x.elf2.str.split("-"))
```

```python
df2
```

```python
def set_maker(x, column):
    return set(range(int(x[column][0]),int(x[column][1])+1))

import functools
elf1_set = functools.partial(set_maker, column = "elf1_parsed")
elf2_set = functools.partial(set_maker,  column = "elf2_parsed")
```

```python
df3 = (
    df2
    [["elf1_parsed", "elf2_parsed"]]
    .assign(elf1_set = lambda x: x.apply(elf1_set, axis = 1), elf2_set = lambda x: x.apply(elf2_set, axis = 1))
)
```

```python
def subset_checker(x, col1: str, col2: str):
    return x[col1].issubset(x[col2])
```

```python
df4 = (
    df3
    .assign(
        col1_sub = df3.apply(lambda x: subset_checker(x, "elf1_set", "elf2_set"), axis =1),
        col2_sub = df3.apply(lambda x: subset_checker(x, "elf2_set", "elf1_set"), axis =1)
    )
)
```

```python
x = df4.col1_sub | df4.col2_sub
```
