```python
df = pd.read_csv("input.txt", sep = " ", names = ['Player1', 'Player2'])
translate = {
    'A': 'Rock',
    'B': 'Paper',
    'C': 'Scissors',
    'X': 'Rock',
    'Y': 'Paper',
    'Z': 'Scissors'
}
df2 = df.assign(Player1 = lambda x: x.Player1.map(translate), 
                Player2 = lambda x: x.Player2.map(translate))
```

```python
df2.head()
```

```python
def calculate_win(series_obj):

    if series_obj[1] == 'Rock':
        return {'Rock': 'Draw', 'Paper': 'Lose', 'Scissors': 'Win'}.get(series_obj[0])
    
    elif series_obj[1] == 'Paper':
        return {'Rock': 'Win', 'Paper': 'Draw', 'Scissors': 'Lose'}.get(series_obj[0])
    
    elif series_obj[1] == 'Scissors':
        return {'Rock': 'Lose', 'Paper': 'Win', 'Scissors': 'Draw'}.get(series_obj[0])
    
    else:
        return "Bad Input"

win_mapper = {'Win': 6, 'Lose': 0, 'Draw': 3}
chosen_tool_mapper = {'Rock': 1, 'Paper': 2, 'Scissors': 3}
```

```python
df3 = (
    df2
    .assign(Outcome = lambda x: x.apply(calculate_win, axis = 1), 
            win_score = lambda x: x.Outcome.map(win_mapper),
            chosen_tool = lambda x: x.Player2.map(chosen_tool_mapper),
            total_score = lambda x: x.win_score + x.chosen_tool)
)
df3
```
